```python
import random

print('Winning rules of the game ROCK PAPER SCISSORS are :\n'
      +"The game has three rounds\n"
      + "Rock vs Paper -> Paper wins \n"
      + "Rock vs Scissors -> Rock wins \n"
      + "Paper vs Scissors -> Scissor wins \n")

# code of the game:

round_number = 3
    
player_score = 0
    
comp_score = 0
    
while round_number > 0 :
    
    print("Enter your choice \n 1 - Rock \n 2 - Paper \n 3 - Scissors \n")
    
    player_choice = int(input("Enter your choice: "))
    
    # If the player enters a value other than the given values, send a warning.
    
    if player_choice>3 or player_choice<1:
        
        print("Please, enter a valid choice")
        
    if player_choice == 1:
        choice_name = "rock"
        
    elif player_choice == 2:
        choice_name = "paper"
        
    else:
        choice_name = "scissors"
        
    print(f"Player choice is {choice_name}\n")
    print('***Now, its computer turn....****\n')
    
    # Next, computer's turn:
    
    comp_choice = random.randint(1,3)
    
    if comp_choice == 1:
        comp_choice_name = "rock"
        
    elif comp_choice == 2:
        comp_choice_name = "paper"
        
    else:
        comp_choice_name = "scissors"
        
    print(f"Computer choice is {comp_choice_name}\n")
              
    # Find the winner:
        
    if player_choice == comp_choice:
            
        print(f"****{choice_name} vs {comp_choice_name}****\n")
        
        player_score = player_score
                
        comp_score = comp_score
        
        round_number =  round_number - 1
      
    elif player_choice == 1:
            
        if comp_choice == 2:
            print(f"****{choice_name} vs {comp_choice_name}****\n")
                
            player_score = player_score
                
            comp_score = comp_score + 1
            
            round_number =  round_number - 1

                    
        elif comp_choice == 3:
            print(f"****{choice_name} vs {comp_choice_name}****\n")
                
            player_score = player_score + 1
                
            comp_score = comp_score 
            
            round_number =  round_number - 1

                    
    elif player_choice == 2:
            
            if comp_choice == 1:
                print(f"****{choice_name} vs {comp_choice_name}****\n")

                player_score = player_score + 1

                comp_score = comp_score
                
                round_number =  round_number - 1

            elif comp_choice == 3:
                print(f"****{choice_name} vs {comp_choice_name}****\n")

                player_score = player_score

                comp_score = comp_score + 1
                
                round_number =  round_number - 1

                    
    elif player_choice == 3:
            
            if comp_choice == 1:
                print(f"****{choice_name} vs {comp_choice_name}****\n")

                player_score = player_score

                comp_score = comp_score + 1
                
                round_number =  round_number - 1

                    
            elif comp_choice == 2:
                print(f"****{choice_name} vs {comp_choice_name}****\n")

                player_score = player_score + 1

                comp_score = comp_score 
                
                round_number =  round_number - 1
                        
                
print(f"Player score: {player_score} and Computer Score {comp_score}")
    
    
if player_score == comp_score:
    print("DRAW")
        
elif player_score > comp_score:
    print("Winner is player!")
    
else:
    print("Winner is computer!")
                
    
    
    
        
        
    

```
